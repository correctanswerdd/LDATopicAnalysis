```python
import sqlite3

name_db = 'papers.db'

def create():
    connection = sqlite3.connect(name_db)
    cursor = connection.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS Paper
                  (arxiv_id TEXT, authors TEXT, title TEXT, abstract TEXT, submit_time TEXT, comment TEXT, cite TEXT)''')

    connection.commit()
    connection.close()


def clear():
    conn = sqlite3.connect(name_db)
    c = conn.cursor()
    c.execute('DELETE FROM Paper;',)
    print('We have deleted', c.rowcount, 'records from the table.')
    conn.commit()
    conn.close()

    conn = sqlite3.connect(name_db)
    c = conn.cursor()
    c.execute("DROP TABLE Paper")
    print('We have deleted table Paper.')
    conn.commit()
    conn.close()


def insert(paper_info):
    conn = sqlite3.connect(name_db)
    c = conn.cursor()
    print('=' * 20)
    print("Opened database successfully")
    print(paper_info)
    try:
        c.execute("INSERT INTO Paper ("
                  "arxiv_id,"
                  "authors,"
                  "title,"
                  "abstract,"
                  "submit_time,"
                  "source) \
              VALUES ({0})".format(str(paper_info).strip('[').strip(']')))
        print("Records created successfully")
    except sqlite3.Error as err:
        print('Query Failed: \nError: %s' % (str(err)))
        pass
    conn.commit()
    conn.close()

clear()
create()
```

```python
import sqlite3
import pandas as pd

path = 'papers0504_platoon.db'
conn = sqlite3.connect(path)

header = []
c = conn.cursor()
for column in c.execute('PRAGMA table_info("Paper")'):
    header.append(column[1])

df = pd.DataFrame(columns=header)
for raw in c.execute('SELECT * FROM Paper'):
    series = pd.Series(list(raw), index=df.columns)
    df = df.append(series, ignore_index=True)

# df.to_csv('abstract0428_powertrain.csv')
conn.close()
```
